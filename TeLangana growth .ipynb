# **Categorize districts into three segments based on their stamp registration revenue generation during the fiscal year 2021 to 2022.**

```python
#filter the year for 2021 and 2022
revenue_2012_2022=Combined_data[(Combined_data['fiscal_year']>=2021)&(Combined_data['fiscal_year']<=2022)]
```

```python
revenue_2012_2022
```

```python
#calcultae total stamp registration
total_doc_rev=(
    revenue_2012_2022
    .groupby('dist_code')['documents_registered_rev']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
```

```python
total_doc_rev.rename(columns={'documents_registered_rev':'stamp_revenue'},inplace=True)
```

```python
total_doc_rev
```

```python
#calculate percentile to define segments
percentiles=total_doc_rev['stamp_revenue'].quantile([0,0.33,0.67,1])
```

```python
percentiles
```

```python
# Create categories based on percentiles

total_doc_rev['segments']=pd.cut(
    total_doc_rev['stamp_revenue'],
    bins=percentiles,
    labels=['low growth','medium growth','high growth'],
    include_lowest=True
)
```

```python
total_doc_rev
```

```python
#merging with districts to get district names
total_rev_districts=pd.merge(total_doc_rev,Districts,on='dist_code',how='left')
```

```python
total_rev_districts
```

```python
# Creating DataFrame
total_rev_districts = pd.DataFrame(total_rev_districts)

# Separate DataFrame into three tables based on 'segments' column
high_segment_df = total_rev_districts[total_rev_districts['segments'] == 'high growth'].sort_values(by='stamp_revenue',ascending=True)
medium_segment_df = total_rev_districts[total_rev_districts['segments'] == 'medium growth']
low_segment_df = total_rev_districts[total_rev_districts['segments'] == 'low growth']

# Print the three tables
print("High Segment DataFrame:")
print(high_segment_df)

print("\nMedium Segment DataFrame:")
print(medium_segment_df)

print("\nLow Segment DataFrame:")
print(low_segment_df)
```
